## 1 절대 경로
* 절대경로는 어떠한 웹페이지나 파일이 가진 고유한 경로로 최상위 디렉토리가 포함된 경로입니다.
* 예) file://User/user/Downloads/
* 예) C:\user\document

## 2 상대 경로
* 상대경로는 현재 위치한 곳을 기준으로 하는 경로입니다.
* / 는 루트
* ./ 는 현재 위치
* ../는 현재 위치의 상위 폴터
* ../../는 두 단계 위의 상위 폴더를 의미합니다.
